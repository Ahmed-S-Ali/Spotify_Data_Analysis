```python
# Calculating Pearson Correlation Matrix for The Selected Features.
graph = df[['track_popularity','danceability', 'energy', 'loudness', 'acousticness', 'instrumentalness',
           'liveness', 'valence', 'tempo', 'time_signature', 'duration_sec']].corr(method = 'pearson')

# Applying a Dark Background Style For Better Visualization.
plt.style.use("dark_background")

# Creating a New Figure For The Heatmap With Specified Size.
plt.figure(figsize=(15,8))

# Generating The Heatmap Using Seaborn's Heatmap Function.
map = sns.heatmap(graph, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='Reds', linewidths=1, linecolor='Black')

# Setting Title For The Heatmap.
map.set_title('Correlation Heatmap between Variable')

# Rotating The x-axis Labels For Better Readability.
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```
